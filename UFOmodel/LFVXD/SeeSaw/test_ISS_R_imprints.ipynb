# Caso Inverse SeeSaw

```python
from Unu_seesaw_ISS_R_imprints import diagonalizationMnu_ISS_svd, Nudata, Mnu
diagonalizationMnu_ISS = diagonalizationMnu_ISS_svd
from mpmath import sqrt, mp, nprint
mp.dps = 80 # Aumentar si se necesita mas precision
d21 = Nudata.squareDm21.central*1e-18
d31 = Nudata.squareDm31.central*1e-18
```

La siguiente definición para comprobar una igualdad númericamente fue tómada del [pep-0485](https://peps.python.org/pep-0485/)

```python
def isclose(a, b, rel_tol=1e-9, abs_tol=0.0):
    return abs(a-b) <= max( rel_tol * max(abs(a), abs(b)), abs_tol )
```

```python
mn1_val = 1e-12
mn2_val, mn3_val = sqrt(mn1_val**2 + d21),sqrt(mn1_val**2 + d31)

MN = 1e4
Mux = 1e-4
thx = 0.1
thy = 0.2
thz = 0.3
mni, UL, UR = diagonalizationMnu_ISS(mn1_val, mn2_val, mn3_val, MN, MN, MN, Mux, Mux, Mux, thx, thy, thz)
mni = sorted(mni)
nprint(mni[0])
nprint(mn1_val)
print(isclose(mni[0], mn1_val, rel_tol=1e-5))
nprint(mni[1])
nprint(mn2_val)
print(isclose(mni[1], mn2_val, rel_tol=1e-5))
nprint(mni[2])
nprint(mn3_val)
print(isclose(mni[2], mn3_val, rel_tol=1e-5))

eps = 1e-3
```

En el caso de neutrinos ligeros tenemos 3 pares de neutrinos cuasidegenerados entre sí, y su diferencia es del orden de $\mu$, como se muestra a continuación

```python
print(f'mu = {Mux}')
print(f'Diferencia mn5 - mn4 = ')
nprint(mni[4] - mni[3])
print(f'Diferencia mn7 - mn6 = ')
nprint(mni[6] - mni[5])
print(f'Diferencia mn9 - mn8 = ')
nprint(mni[8] - mni[7])
```

La forma de la matriz está dada por

```python
MN = 1e4
Mux = 1e-4
thx = 0.1
thy = 0.2
thz = 0.3
nprint(Mnu(mn1_val, mn2_val, mn3_val, MN, MN, MN, Mux, Mux, Mux, thx, thy, thz))
```

```python
def eigenvalores_masas_son_cercanas(MN, Mux, thx, thy, thz, eps=1e-3):
    mni, UL, UR = diagonalizationMnu_ISS(mn1_val, mn2_val, mn3_val, MN, MN, MN, Mux, Mux, Mux, thx, thy, thz)
    mni = sorted(mni)
    return all([
        isclose(mni[0], mn1_val, rel_tol=eps), 
        isclose(mni[1], mn2_val, rel_tol=eps),
        isclose(mni[2], mn3_val, rel_tol=eps),
        isclose(mni[3], MN, rel_tol=eps),
        isclose(mni[4], MN, rel_tol=eps),
        isclose(mni[5], MN, rel_tol=eps),
        isclose(mni[6], MN, rel_tol=eps),
        isclose(mni[7], MN, rel_tol=eps),
        isclose(mni[8], MN, rel_tol=eps)
    ])
```

```python
eigenvalores_masas_son_cercanas(1e4, 1e-6, 0.1, 0.2, 0.3)
```

Se prueba que los eigenvalores sean 

```python
bool_list = []
for M in [1e3, 1e4, 1e5]:
    for mux in [1e-5, 1e-4, 1e-3]:
        for th1 in [0.1, 0.3, 0.5, 0.7, 0.9]:
            for th2 in [0.1, 0.3, 0.5, 0.7, 0.9]:
                for th3 in [0.1, 0.3, 0.5, 0.7, 0.9]:
                    bool_list.append(eigenvalores_masas_son_cercanas(M, mux, th1, th2, th3))
all(bool_list)
```
